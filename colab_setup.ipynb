## How to handle your GitHub token SAFELY

**Do not paste the token into a notebook cell.** Anything in a notebook cell
is saved with the notebook and visible to anyone you share it with.

Instead, use Colab secrets:
1. In the Colab sidebar, click the key icon ("Secrets")
2. Click "Add new secret"
3. Name: `GITHUB_TOKEN`, value: your fine-grained token
4. Toggle "Notebook access" ON
